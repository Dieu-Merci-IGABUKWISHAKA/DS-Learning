```python
datasummary = (
    data.groupby(["Country", "Year"])["Total Energy Consumption (TWh)"]
    .sum()
    .reset_index()
)
```

```python
datasummary
```

```python
country_name = "Canada"
country_data=datasummary[datasummary['Country']==country_name]
```

```python
plt.figure(figsize=(10,6))
plt.plot(country_data['Year'],country_data['Total Energy Consumption (TWh)'])
plt.xlabel("Year")
plt.ylabel("Total Energy Consumption (TWh)")
plt.title(f"Energy Consumption Trend - {country_name}")
plt.show()
```
